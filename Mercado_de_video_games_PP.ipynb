Exploração Inicial e correção de dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/vgsales.csv')
```

```python
df.head()
```

```python
# Dicionário para traduzir as colunas
traducao_colunas = {
    'Rank': 'Ranking geral de vendas',
    'Name': 'Nome do jogo',
    'Platform': 'Console/Plataforma',
    'Year': 'Ano de lançamento',
    'Genre': 'Gênero do jogo',
    'Publisher': 'Empresa que publicou o jogo',
    'NA_Sales': 'Vendas na América do Norte (em milhões)',
    'EU_Sales': 'Vendas na Europa (em milhões)',
    'JP_Sales': 'Vendas no Japão (em milhões)',
    'Other_Sales': 'Vendas em outras regiões (em milhões)',
    'Global_Sales': 'Vendas totais no mundo (em milhões)'
}

# Renomear as colunas
df.rename(columns=traducao_colunas, inplace=True)

# Exibir os primeiros registros para verificar as novas colunas
display(df.head())
```

```python
df['Gênero do jogo'] = df['Gênero do jogo'].astype(str)
display(df.head())
```

# Fase 03 - Análise Profunda

Qual gênero vende mais em cada região (América do Norte, Europa e Japão)? Existem diferenças culturais claras?

```python
# Group by genre and sum sales for each region
regional_genre_sales = df.groupby('Gênero do jogo')[['Vendas na América do Norte (em milhões)', 'Vendas na Europa (em milhões)', 'Vendas no Japão (em milhões)']].sum()

# Find the best-selling genre in each region
best_genre_na = regional_genre_sales['Vendas na América do Norte (em milhões)'].idxmax()
best_genre_eu = regional_genre_sales['Vendas na Europa (em milhões)'].idxmax()
best_genre_jp = regional_genre_sales['Vendas no Japão (em milhões)'].idxmax()

print(f"Gênero mais vendido na América do Norte: {best_genre_na}")
print(f"Gênero mais vendido na Europa: {best_genre_eu}")
print(f"Gênero mais vendido no Japão: {best_genre_jp}")

# Display the regional sales by genre
display(regional_genre_sales)
```

Qual foi o "auge" e o "declínio" de vendas de plataformas famosas como 'Wii', 'PS2' e 'X360'?


```python
import plotly.express as px

platforms = ['Wii', 'PS2', 'X360']

for platform in platforms:
    # Filter data for the current platform
    platform_df = df[df['Console/Plataforma'] == platform]

    # Group by year and sum global sales
    annual_sales = platform_df.groupby('Ano de lançamento')['Vendas totais no mundo (em milhões)'].sum()

    # Create an interactive line plot
    fig = px.line(x=annual_sales.index, y=annual_sales.values,
                  labels={'x': 'Ano de lançamento', 'y': 'Vendas totais no mundo (em milhões)'},
                  title=f'Vendas Anuais para a Plataforma: {platform}')
    fig.show()
```

Onde está a "mina de ouro"? Qual gênero vendeu melhor em qual plataforma?

```python
# Get top 5 genres and top 5 platforms
top_5_genres = df.groupby('Gênero do jogo')['Vendas totais no mundo (em milhões)'].sum().sort_values(ascending=False).head(5).index.tolist()
top_5_platforms = df.groupby('Console/Plataforma')['Vendas totais no mundo (em milhões)'].sum().sort_values(ascending=False).head(5).index.tolist()

# Filter the dataframe to include only the top genres and platforms
filtered_df = df[df['Gênero do jogo'].isin(top_5_genres) & df['Console/Plataforma'].isin(top_5_platforms)]

# Create a pivot table
pivot_table = pd.pivot_table(filtered_df, values='Vendas totais no mundo (em milhões)', index='Gênero do jogo', columns='Console/Plataforma', aggfunc='sum')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Vendas Totais por Gênero e Plataforma (Top 5)')
plt.xlabel('Console/Plataforma')
plt.ylabel('Gênero do jogo')
plt.show()
```

## Resumo e Recomendações

A partir da análise das vendas de jogos, alguns pontos se destacam:

* **Gêneros e Plataformas em Alta:** Ação, Esportes e RPG se mostram os gêneros mais fortes. Já plataformas como PS2, DS, PS3, Wii e X360 lideraram em volume de vendas.
* **Preferências Regionais:** Enquanto Ação domina nos EUA e Europa, no Japão os jogadores preferem RPGs.
* **Ciclo das Plataformas:** Cada console tem seu auge e queda, como vimos no Wii, PS2 e X360. Esse ciclo é crucial para planejar lançamentos.
* **Combinações Vencedoras:** O cruzamento entre gênero e plataforma revelou “minas de ouro”, como Ação no PS3 e X360, e Esportes no Wii.

Com base nisso, seguem 5 recomendações práticas para estúdios de jogos:

1. **Aposte nos gêneros e consoles mais fortes:** Invista em Ação e Esportes nas plataformas com maior histórico de vendas.
2. **Respeite o gosto local:** RPGs tendem a ter mais sucesso no Japão, enquanto no Ocidente Ação e Esportes seguem no topo.
3. **Fique de olho no ciclo de vida dos consoles:** Planeje lançamentos alinhados aos momentos de maior popularidade de cada plataforma.
4. **Explore nichos promissores:** Além dos grandes gêneros, avalie combinações de nicho que tenham potencial e menos concorrência.
5. **Aprenda com os grandes:** Estude publishers como Nintendo, EA e Activision para entender práticas que podem ser adaptadas ao seu estúdio.
